```python
#1.1
# sentence count 
import pandas as pd
data_url = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
dataframe = pd.read_csv(data_url,sep='delimiter',engine ='python', header=None)
dataframe['sentence_count'] = dataframe[0].apply(lambda x:len(str(x).split(".")))
dataframe
```

```python
#  Word Count 
dataframe['word_count'] = dataframe[0].apply(lambda x: len(str(x).split(" ")))
dataframe
```

```python
# Character count
dataframe['char_count'] = dataframe[0].str.len()
print("\nCharacter Count:")
dataframe
```

```python
#Avg word
print("\nAverage Word Length:")
def avg_word(sentence):
  words = str(sentence).split()
  return (sum(len(word) for word in words)/len(words))

dataframe['avg_word'] = dataframe[0].apply(lambda x: avg_word(x))
dataframe
```

```python
#stop words
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
dataframe['stopwords'] = dataframe[0].apply(lambda x: len([x for x in x.split() if x in stop]))
dataframe
```

```python
#special character
import re
dataframe['special_chars'] = dataframe[0].apply(lambda x: sum(not (x.isalpha() and x.isdigit()) for y in x ))
dataframe
```

```python
#numerics
dataframe['numerics'] = dataframe[0].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
dataframe
```

```python
#uppercase
dataframe['upper'] = dataframe[0].apply(lambda x: len([x for x in x.split() if x.isupper()]))
dataframe
```

```python
#1.2
#lower case
dataframe["lower_case"] = dataframe[0].apply(lambda x: " ".join(x.lower() for x in x.split()))
dataframe
```

```python
#punctuation removal
dataframe["punctuationremoval"] = dataframe["lower_case"].str.replace('[^\w\s]','')
dataframe
```

```python
#stop words removal
from nltk.corpus import stopwords
stop = stopwords.words('english')
dataframe["stopwords"] = dataframe["punctuationremoval"].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
dataframe
```

```python
#commonwords removal
freq = pd.Series(' '.join(dataframe[0]).split()).value_counts()[:10]#10commonword
freq
freq = list(freq.index)#remove above commonwords
dataframe["commonwords"] = dataframe["stopwords"].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
dataframe
```

```python
#remove rarewords
freq = pd.Series(' '.join(dataframe[0]).split()).value_counts()[-10:]#rarewords list
freq
freq = list(freq.index)
dataframe["rarewords"] = dataframe["commonwords"].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
dataframe
```

```python
#spelling correction
from textblob import TextBlob
dataframe["spell_correction"] = dataframe["rarewords"].apply(lambda x: str(TextBlob(x).correct()))
dataframe
```

```python
#tokenization
import nltk
nltk.download('punkt')
dataframe["tokenization"] = dataframe["spell_correction"].apply(lambda x: TextBlob(x).words)
dataframe
```

```python
#stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
dataframe["stemming"]= dataframe["tokenization"].apply(lambda x: " ".join([st.stem(word) for word in x]))
dataframe
```

```python

#lemmatization
from textblob import Word
import nltk
nltk.download('wordnet')
dataframe["lemmatization"] = dataframe["stemming"].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
dataframe
```

```python
#1.3
print("\nSaving to CSV file")
dataframe.to_csv("resultfile.csv")
```

```python
#1.4
#term frequency
tf1 = (dataframe["lemmatization"]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
```

```python
#n-gram
import collections

def ngramming(nval):
  dataframe[str(nval)+"gram"] = dataframe["lemmatization"].apply(lambda x: [' '.join(y) for y in nltk.ngrams(nltk.word_tokenize(x), nval)])
  #print(dataframe[str(nval)+"gram"])
  newlist =[]
  for i in dataframe[str(nval)+"gram"]:
    newlist = newlist + i
  a = collections.Counter(newlist).most_common(10)
  for l in a:
    print(l[0],l[1])

for i in range(1,4):
  print(i,"-gram results:")
  ngramming(i)
  print("\n")
```
