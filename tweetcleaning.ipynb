```python
#duplicates
import json
import csv
import pandas as pd
df = pd.read_json("/content/tweetcollectionhashtags.json")
print(df.head())

a = 0

l = []
for i in df["tweets"]:
  ci = df["cui"][a]
  conc = df["concepts"][a]
  for j in i:
    k = {}
    k["cui"] = ci
    k["concepts"] = conc
    k["tweets"] = j
    l.append(k)
  a = a+1
df2 = pd.DataFrame(l)
```

```python
#Collecting Hashtags
import json
import csv
import pandas as pd
import re


hastags = []

for i in df2["tweets"]:
  a= re.findall(r"#(\w+)", str(i))
  b = ", ".join(a)
  hastags.append(b)

df2["hashtags"]=hastags
df2.to_csv("tweetcollection.csv")

#hashtags frequency
from collections import Counter
Counter(" ".join(hastags).split(" ")).items()
```

```python
#Cleaning

import json
import csv
import pandas as pd
import requests
from textblob import Word
import nltk
nltk.download('wordnet')
nltk.download('punkt')
from textblob import TextBlob
from nltk.stem import PorterStemmer
tweets = []
cleanedtweets = []
df2['linksremoved'] = df2['tweets'].str.replace('http\S+|www.\S+', '', case=False)
df2["punctuationremoved"] = df2["linksremoved"].str.replace('[^\w\s]','')#punctuation
df2["lowercased"] = df2["punctuationremoved"].apply(lambda x: " ".join(x.lower() for x in x.split()))#lowercase
df2["spellings"] = df2["lowercased"].apply(lambda x: str(TextBlob(x).correct()))#spellings
df2.to_csv("Cleanedtweets.csv")
```

```python
import re
import pandas as pd
df = pd.read_csv("/content/Cleanedtweets.csv")
#df.head(5) # with emoji
cleaned = []
for line in df["spellings"]:
  emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
  cleaned.append(emoji_pattern.sub(r'', line))
df["cleanedtweets"] = cleaned



df.head(5) # with emoji
df.to_csv("Cleaned_tweets.csv")
```
